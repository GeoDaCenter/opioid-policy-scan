```python
import PyPDF2
```

```python
pdfFileObj = open('directory_2017.pdf', 'rb')
pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
pdfReader.numPages
```

## Single Page extraction

```python
# data starts from page# 15
page1 = pdfReader.getPage(14).extractText()
rawdata = page1.split('\n')
rawdata
```

```python
# remove NA lines in rawdata
data = []
for line in rawdata:
    if line == '' or line == 'KEY':
        continue
    data.append(line)
data
```

```python
# total_center_info_list = [[center#1_info], [center#2_info] ...]
total_center_info_list = []
center_temp = [data[0]]
pre = data[0]
# How we detect the start of new center info:
# when the previous line is all cap, i.e. the codes of the last center
# and the current line is mixed
# this indicates the current line is the first line of a new center
for entry in data[1:]:
    if pre.isupper() and (not entry.isupper()):
        total_center_info_list.append(center_temp)
        center_temp = [entry]
    else:
        center_temp.append(entry)
    pre = entry
total_center_info_list.append(center_temp)
```

## Turn list into dictionary for single center

```python
test = total_center_info_list[4]
test
```

```python
def list_to_dict(test: list):
    name = []
    address = []
    phone = []
    codes = []
    for i in range(len(test)):
        entry = test[i]
        if entry[0] == ' ':
            entry = entry[1:]
        if entry[0].isdigit():
            name = test[:i]
            address_idx = i
            # print('address line: {}'.format(address_idx))
        elif entry[0] == '(':
            address = test[address_idx:i]
            phone_idx = i
            # print('phone line: {}'.format(phone_idx))
        elif entry.isupper() and (not test[i-1].isupper()):
            phone = test[phone_idx:i]
            codes_idx = i
            codes = test[i:]
    # or we can make each center a dictionary for simplicity
    res = {"name": name, "address": address,
           "phone": phone, "codes": codes}
    return res
```

```python
centers = []
for center in total_center_info_list:
    centers.append(list_to_dict(center))
centers
```

```python
address_list = [center['address'] for center in centers]
address_list
```
